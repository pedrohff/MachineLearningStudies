```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
filename = os.path.join('iris.csv')
```

```python
!head iris.csv
```

```python
dados = np.loadtxt(filename, delimiter=',', usecols=(0,1,2,3),skiprows=1)
nomeColunas = np.loadtxt(filename, delimiter=',', usecols=(0,1,2,3), dtype=str)[0]
nomeColunas
```

```python
sepal_length = []
for x in range(0,len(dados)):
    sepal_length.append(dados[x,0])

sepal_length = np.array(sepal_length)
sepal_length[:5]
```

```python
sepal_width = []
for x in range(0,len(dados)):
    sepal_width.append(dados[x,1])

sepal_width = np.array(sepal_width)
sepal_width[:5]
```

```python
arquivocompleto = np.loadtxt(filename, delimiter=',',skiprows=1, dtype=str)
aux = arquivocompleto.tolist()
for rosa in aux:
    for x in range(0,len(rosa)-1):
        rosa[x] = float(rosa[x])

aux
```

```python
setosa = []
for x in range(0,len(arquivocompleto)):
    if arquivocompleto[x,4] == 'setosa':
        setosa.append(arquivocompleto[x][0:4])

setosa = np.array(setosa,dtype=float)
setosa
```

```python
versicolor = []
for x in range(0,len(arquivocompleto)):
    if arquivocompleto[x,4] == 'versicolor':
        versicolor.append(arquivocompleto[x][0:4])

versicolor = np.array(versicolor,dtype=float)
versicolor
```

# Sepal Length - Setosa x Versicolor

```python
#['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
vSL=[]
sSL=[]

for x in versicolor:
    vSL.append(x[0])
    
for x in setosa:
    sSL.append(x[0])
    

plt.scatter(vSL, sSL,c=['k','r'])


plt.show()
```

# Sepal Width - Setosa x Versicolor

```python
#['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
vSW=[]
sSW=[]

for x in versicolor:
    vSW.append(x[1])
    
for x in setosa:
    sSW.append(x[1])
    

plt.scatter(vSW, sSW,c=['k','r'])
plt.show()
```

# Petal Length - Setosa x Versicolor

```python
#['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
vPL=[]
sPL=[]

for x in versicolor:
    vPL.append(x[2])
    
for x in setosa:
    sPL.append(x[2])
    

plt.scatter(vPL, sPL,c=['k','r'])
plt.show()
```

# Petal Width - Setosa x Versicolor

```python
#['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
vPW=[]
sPW=[]

for x in versicolor:
    vPW.append(x[3])
    
for x in setosa:
    sPW.append(x[3])
    

plt.scatter(vPW, sPW,c=['k','r'])
plt.show()
```
